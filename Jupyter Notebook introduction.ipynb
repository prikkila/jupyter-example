## Example of plotting exercise:

Code below makes a plot of function $y = \sin(x)$. Try to execute the code by running the cell below. What happens?

```python
# Let's import a scientific calculating module called "numpy".
import numpy as np
# Let's import a module called "matplotlib" for plotting things.
import matplotlib.pyplot as plt
# Let's import some more modules.
from scipy.stats import norm
import matplotlib.mlab as mlab


# Let's start with creating an example dataset.

# This creates an array of evenly spaced number over the selected interval. The first argument is the start of the
# interval and the second argument is the end of the interval. The third argument sets the amount of evenly spaced numbers.
x = np.linspace(0,20,20)

# This creates an array with numbers sin(x1), sin(x2), sin(x3), ... on it, where x = [x1, x2, x3, ...] from the rox below. 
y = np.sin(x) 


# Jupyter Notebook has predefined "magic functions". This one is for making the plot visible straight in notebook.
%matplotlib inline

# Plot details and making of plot.
plt.title("An example plot. Not so accurate.")
plt.xlabel("Name of the x-label")
plt.ylabel("Name of the y-label")
plt.plot(x,y)
```

## Example of histogram and fitting

Let's make a simple histogram from random data we create. Let's also make a fit on the histogram.
<br>
<br>

```python
# Let's create a sample of "standard normal" distribution of 1000 numbers with numpy-package.
random_numbers = np.random.randn(1000)
```

```python
# This makes a histogram from the distribution we created.
n,bins,patches = plt.hist(random_numbers,20,normed=True,facecolor='burlywood')

# This makes a fit to the distribution.
(mu, sigma) = norm.fit(random_numbers)
plt.figure(1)
y = mlab.normpdf(bins,mu,sigma)
plt.plot(bins,y,'r--',linewidth=2)

# Plot details.
plt.title("A historgam example")
plt.xlabel("Value")
plt.ylabel("Frequency")
```
